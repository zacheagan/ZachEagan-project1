```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('music_data.csv')
df.head()
```

```python
region = df.groupby('Region').mean()
region.plot.bar(rot=75)
plt.ylabel('Average of each variable')
plt.title("Region vs. Integer data")
```

```python
major = df.groupby('Major').mean()
major.plot.bar(rot=75)
plt.ylabel('Average of each variable')
plt.title("Major vs. Integer data")
```

```python
music_genre = df.groupby('Favorite Music Genre').mean()
music_genre.plot.bar(rot=75)
plt.ylabel('Average of each variable')
plt.title("Favorite Music Genre vs. Integer data")
```
