```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.describe()
```

```python
iris.hist('sepal_length')
plt.ylabel('number of observations')
```

```python
species_avg = iris.groupby('species').mean()
```

```python
species_avg.plot.bar()
```

```python
print(iris[['petal_length', 'petal_width']].corr())
iris.plot.scatter(x='petal_length', y='petal_width')
```

```python
sns.scatterplot(x='petal_length', y='petal_width', data=iris, hue='species')
```

```python
iris = iris.sort_values(['petal_length'], ascending=True)
```

```python
iris.plot.line(x='petal_length', y='petal_width')
```

```python
species_pct = iris.species.value_counts(normalize=True)
```

```python
species_pct.plot.pie()
plt.legend(loc='right')
```
